```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
import matplotlib as mpl
plt.style.use('science')
mpl.rcParams['font.size'] = 15

# Set major tick size
mpl.rcParams['xtick.major.size'] = 0
mpl.rcParams['ytick.major.size'] = 0

# Set minor tick size
mpl.rcParams['xtick.minor.size'] = 0
mpl.rcParams['ytick.minor.size'] = 0
```

```python
otoc_tebd = np.loadtxt('OTOCLightcone/tebd_nn_lightcone.txt')
otoc_tdvp_nn = np.loadtxt('OTOCLightcone/otoc_lightcone_nn.txt')
otoc_tdvp_nnn = np.loadtxt('OTOCLightcone/otoc_lightcone_nnn.txt')
otoc_tdvp_nnnn = np.loadtxt('OTOCLightcone/otoc_lightcone_nnnn.txt')
```

```python
plt.figure(figsize=(5, 4))
plt.imshow(otoc_tebd, aspect='auto', cmap='magma', origin='lower')
plt.colorbar(label=r'$C(r, t)$')
plt.xlabel(r'$r$')
plt.ylabel(r'$t$')
plt.xticks([0, 4, 8, 12, 16, 20, 24, 28, 32], 
           [r'$-16$', r'$-12$', r'$-8$', r'$-4$', r'$0$', r'$4$', r'$8$', r'$12$', r'$16$'],
           fontsize=12)
plt.yticks([1, 3, 5, 7, 9, 11, 13], [r'$1$', r'$2$', r'$3$', r'$4$', r'$5$', r'$6$', r'$7$'])
plt.savefig('figures/tebd_nn_lightcone.pdf')
```

```python
plt.figure(figsize=(5, 4))
plt.imshow(otoc_tdvp_nn, aspect='auto', cmap='magma', origin='lower')
plt.colorbar(label=r'$C(r, t)$')
plt.xlabel(r'$r$')
plt.ylabel(r'$t$')
plt.xticks([0, 4, 8, 12, 16, 20, 24, 28, 32], 
           [r'$-16$', r'$-12$', r'$-8$', r'$-4$', r'$0$', r'$4$', r'$8$', r'$12$', r'$16$'],
           fontsize=12)
plt.yticks([1, 2, 3, 4], [r'$0.5$', r'$1$', r'$1.5$', r'$2$'])
plt.savefig('figures/tvdp_nn_lightcone.pdf')
```
