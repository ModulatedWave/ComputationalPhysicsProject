```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
import matplotlib as mpl
plt.style.use('science')

mpl.rcParams['font.size'] = 15

# Set major tick size
mpl.rcParams['xtick.major.size'] = 0
mpl.rcParams['ytick.major.size'] = 0

# Set minor tick size
mpl.rcParams['xtick.minor.size'] = 0
mpl.rcParams['ytick.minor.size'] = 0
```

```python
K_2 = np.loadtxt('KCA_data/d_K_2_N_2048_t_400_n_100.txt')
K_3 = np.loadtxt('KCA_data/d_K_3_N_2048_t_400_n_100.txt')
K_4 = np.loadtxt('KCA_data/d_K_4_N_2048_t_400_n_500.txt')
K_4_frozen = np.loadtxt('KCA_data/d_K_4_N_2048_t_400_n_100_p=0.11.txt')
K_5 = np.loadtxt('KCA_data/d_K_5_N_2048_t_400_n_25.txt')
K_6 = np.loadtxt('KCA_data/d_K_6_N_2048_t_400_n_20.txt')
```

```python
K_2 = np.where(K_2 > 0.5, 0.5, K_2)
K_3 = np.where(K_3 > 0.5, 0.5, K_3)
K_4 = np.where(K_4 > 0.5, 0.5, K_4)
K_4_frozen = np.where(K_4_frozen > 0.5, 0.5, K_4_frozen)
K_5 = np.where(K_5 > 0.5, 0.5, K_5)
K_6 = np.where(K_6 > 0.5, 0.5, K_6)
```

```python
plt.figure(figsize=(5, 4))
plt.imshow(K_4, aspect='auto', cmap='Blues', origin='lower')
plt.colorbar(label=r'$d(x, t)$')
plt.xlabel(r'$r$', fontsize=15)
plt.ylabel(r'$t$', fontsize=15)
plt.title(r'$p=0.40$', fontsize=15)
plt.xticks([0, 512, 1024, 1536, 2048], [r'$-1000$', r'$-500$', r'$0$', r'$500$', r'$1000$'], fontsize=15)
plt.savefig('figures/K_4.pdf')
```
